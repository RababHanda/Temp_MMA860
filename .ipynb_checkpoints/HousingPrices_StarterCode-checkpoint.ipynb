#### Data Cleaning

```python
#checking for datatype
print(df_test.dtypes)
```

```python
#checking for nulls
cols = df_test.columns.tolist()

for col in cols:
    null_count = df_test[col].isnull().sum()
    if null_count > 0:
        print(f"{col}: {null_count}")
```

```python
#check for duplicates
duplicate_count = df_test.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
```

```python
#Checking for distinct values in all colums of datatype object
num_cols = df_test.select_dtypes(include=['number']).columns.tolist()

# Box plot for each numeric column
for col in num_cols:
    plt.figure(figsize=(5, 4))
    sns.boxplot(x=df_test[col])
    plt.title(f'Boxplot for {col}')
    plt.show()
```
